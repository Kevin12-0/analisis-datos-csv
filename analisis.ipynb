```python
df = pd.DataFrame(df_copy,columns=['Victim_Category','Victims_Age'])
df['Cantidad'] = ''
df
```

```python
df2 = pd.DataFrame.groupby(df, by=['Victim_Category','Victims_Age'])
cantidad_category_age = df2.count()
```

```python
cantidad_category_age
```

```python
df1= cantidad_category_age['Cantidad']
df1_occupant = df1['3rd Party driver / Occupant']
plt.barh(df1_occupant.index,df1_occupant.values)
plt.ylabel("Edad de usuarios")
plt.xlabel("Total de victimas")
plt.title("Grafica de victimas Occupant")
plt.show()
```

```python
df3= cantidad_category_age['Cantidad']
df3_occupant = df3['Passenger']
plt.barh(df3_occupant.index,df3_occupant.values)
plt.ylabel("Edad de usuarios")
plt.xlabel("Total de victimas")
plt.title("Grafica de victimas Passenger")
plt.show()
```
